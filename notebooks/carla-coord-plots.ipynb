```python
path = Path('/ubc/cs/research/plai-scratch/video-diffusion-shared/datasets/train_test_splits/carla-no-traffic')

def read_coords(csv_name):
    df = pd.read_csv(path / csv_name, index_col=0)
    return np.stack(df.path.apply(np.load))

def reshape_np(arr):
    return arr.reshape(arr.shape[0]*arr.shape[1], arr.shape[2])

# B, T, C
coords_train = read_coords('coords_train.csv')
coords_test = read_coords('coords_test.csv')
```

```python
train_xy = reshape_np(coords_train)[:, [0,1]]
test_xy = reshape_np(coords_test)[:, [0,1]]
```

```python
ax = sns.lineplot(x=train_xy[:,0], y=train_xy[:,1], sort=False, lw=1)
```
